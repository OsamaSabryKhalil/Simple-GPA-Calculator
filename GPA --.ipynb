```python
def GPA(Subjects_number):    
    import numpy  as np
    Subject_Name_list   = []
    Subject_Grade_list  = []
    Subject_Point_list  = []
    Subject_Credit_list = []
    i=1
    while i <= Subjects_number :
        i = i+1
        sub = input(    "Enter the Subject's name\n")
        Subject_Name_list.append(sub)
        gra = input(    "Enter the Subject's grade\n")
        Subject_Grade_list.append(gra)
        if   gra == 'A+':
            Subject_Point_list.append(4.0)
        elif gra =='A':
            Subject_Point_list.append(4.0)
        elif gra == 'A-':
            Subject_Point_list.append(3.7)
        elif gra == 'B+':
            Subject_Point_list.append(3.3)
        elif gra == 'B' :
            Subject_Point_list.append(3.0)
        elif gra == 'B-':
            Subject_Point_list.append(2.7)
        elif gra == 'C+':
            Subject_Point_list.append(2.3)
        elif gra == 'C' :
            Subject_Point_list.append(2.0)
        elif gra == 'C-':
            Subject_Point_list.append(1.7)
        elif gra == 'D+':
            Subject_Point_list.append(1.3)
        elif gra == 'D' :
            Subject_Point_list.append(1.0)
        elif gra == 'D-':
            Subject_Point_list.append(0.7)
        elif gra == 'F':
            Subject_Point_list.append(0.0)
        cre = int(input('Enter the credit for each subject\n')) 
        if gra == 'D-':
            Subject_Credit_list.append(0)
        elif gra =='F':
            Subject_Credit_list.append(0)
        else:
            Subject_Credit_list.append(cre)



    Subject_Name_list   = np.array(Subject_Name_list)
    Subject_Grade_list  = np.array(Subject_Grade_list)
    Subject_Point_list  = np.array(Subject_Point_list)
    Subject_Credit_list = np.array(Subject_Credit_list)
    grad_test = (Subject_Point_list) * (Subject_Credit_list) 
    GPA = grad_test/np.sum(Subject_Credit_list)
    print('Your GPA >> ',np.sum(GPA))       


            

            
        
    
    
```
